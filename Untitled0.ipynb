```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/marketing_data.csv")
display(df.head())
```

#check data type

```python
print("Columns in DataFrame:", df.columns.tolist())
print(df[['Dt_Customer', ' Income ']].dtypes)
```

#validate Dt_customer

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], errors='coerce')

print("Date range:", df['Dt_Customer'].min(), "to", df['Dt_Customer'].max())
print("Missing Dt_Customer:", df['Dt_Customer'].isnull().sum())

```

validate Income

```python
df[' Income '] = df[' Income '].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)
print(df[' Income '].describe())
print("Missing Income:", df[' Income '].isnull().sum())
print("Negative Income values:", (df[' Income '] < 0).sum())
```

#2 clean education data

```python
df['Education'] = df['Education'].str.strip().str.title()

education_map = {
    'Graduation': 'Graduate',
    '2Nd Cycle': 'Graduate',
    'Basic': 'Basic',
    'Phd': 'PhD',
    'Master': 'Master'
}

df['Education'] = df['Education'].replace(education_map)

```

#clean marital status

```python
df['Marital_Status'] = df['Marital_Status'].str.strip().str.title()

marital_map = {
    'Married': 'Married',
    'Together': 'Married',   # Often grouped as similar household structure
    'Single': 'Single',
    'Alone': 'Single',
    'Divorced': 'Divorced',
    'Widow': 'Widowed',
    'Widower': 'Widowed',
    'Absurd': 'Other',
    'YoLo': 'Other'
}

df['Marital_Status'] = df['Marital_Status'].replace(marital_map)

```

#Compute group means

```python
group_means = df.groupby(['Education', 'Marital_Status'])[' Income '].transform('mean')
```

#Input missing values

```python
df[' Income '] = df[' Income '].fillna(group_means)
```

```python
df[' Income '] = df[' Income '].fillna(df[' Income '].median())
```

#Validate the inputation

```python
print("Remaining missing Income:", df[' Income '].isnull().sum())
print(df[['Education', 'Marital_Status', ' Income ']].head())
```
